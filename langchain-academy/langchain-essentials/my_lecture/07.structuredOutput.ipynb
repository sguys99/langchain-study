## Structured Output Example

```python
from typing_extensions import TypedDict

from langchain.agents import create_agent


class ContactInfo(TypedDict):
    name: str
    email: str
    phone: str


agent = create_agent(model="openai:gpt-5-mini", response_format=ContactInfo)

recorded_conversation = """존 도우 씨와 통화했습니다. 그는 예시 회사에서 근무합니다. 그의 전화번호는, 보자, 555-1234567입니다. 잘 적으셨나요?
그리고 이메일은 john@example.com입니다. 그는 캡틴 크런치 시리얼 50박스를 주문하고 싶어 했습니다."""

result = agent.invoke(
    {"messages": [{"role": "user", "content": recorded_conversation}]}
)

result["structured_response"]
```

#### Multiple data types are supported 

* pydantic `BaseModel`
* `TypedDict`
* `dataclasses`
* json schema (dict)

```python
from langchain.agents import create_agent
from pydantic import BaseModel


class ContactInfo(BaseModel):
    name: str
    email: str
    phone: str


agent = create_agent(model="openai:gpt-5-mini", response_format=ContactInfo)

recorded_conversation = """ 존 도우 씨와 통화했습니다. 그는 예시 회사에서 근무합니다. 그의 전화번호는, 보자, 555-1234567입니다. 잘 적으셨나요?
그리고 이메일은 john@example.com입니다. 그는 캡틴 크런치 시리얼 50박스를 주문하고 싶어 했습니다."""

result = agent.invoke(
    {"messages": [{"role": "user", "content": recorded_conversation}]}
)

result["structured_response"]
```
