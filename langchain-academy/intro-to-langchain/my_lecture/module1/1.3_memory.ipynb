## No memory

```python
from langchain.agents import create_agent


agent = create_agent(
    "gpt-5-nano"
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="내 이름은 유광명이고 회색을 좋아해.")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="내가 좋아하는 색이 뭐야?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    "gpt-5-nano",
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="내이름은 유광명이고 나는 회색을 좋아해")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="내가 좋아하는 색이 뭐야?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```
