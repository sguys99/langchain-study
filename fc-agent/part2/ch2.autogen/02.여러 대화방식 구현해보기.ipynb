```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from autogen import ConversableAgent
```

```python
coder_agent = ConversableAgent(
    name="Junior_Coder_Agent",
    system_message="당신은 3년차 파이썬 코드 전문 소프트웨어 엔지니어입니다. 이해가 안되는 부분은 Senior Coder Agent에게 물어보세요.",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
)

manager_agent = ConversableAgent(
    name="Seinor_Coder_Agent",
    system_message="""당신은 20년차 파이썬 코드 전문 소프트웨어 엔지니어입니다.
    질문이 주어지면, 해당 질문에 전문성을 가지고 답해주세요.
    만약 코드가 주어졌다면 주어진 코드를 검토하고 효율성을 높일 방안을 탐구하세요.
    """,
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
)
```

```python
chat_result = coder_agent.initiate_chat(
    manager_agent,
    message="피보나치 수열 코드에 대해서 한글로 설명해주세요",
    summary_method="reflection_with_llm",
    max_turns=2,
)
```

```python
print(chat_result.summary)
```

### 순차적 대화 방식

```python
topic_agent = ConversableAgent(
    name="topic_agent",  # 영어로 작성, 띄워쓰기 없음
    system_message="당신은 토론할 주제를 제시하는 역할입니다. 현재 사회적으로 중요한 주제를 하나 선정하여 제시해 주세요",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
    human_input_mode="NEVER",  # 사람의 개입 없음
)

economic_agent = ConversableAgent(
    name="economic_agent",
    system_message="당신은 경제학자입니다. 주어진 주제에 대해 경제적 관점에 대해 의견을 제시해 주세요",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
    human_input_mode="NEVER",
)

social_agent = ConversableAgent(
    name="social_agent",
    system_message="당신은 사회학자입니다. 주어진 주제와 이전 의견들을 고려해서 사회적 관점에서 의견을 제시해 주세요",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
    human_input_mode="NEVER",
)

environmental_agent = ConversableAgent(
    name="environmental_agent",
    system_message="당신은 환경 전문가입니다. 주어진 주제와 이전 의견들을 고려해서 환경적 관점에서 의견을 제시해 주세요",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
    human_input_mode="NEVER",
)

ethical_agent = ConversableAgent(
    name="ethical_agent",
    system_message="당신은 윤리학자입니다. 주어진 주제와 이전 의견들을 고려해서 윤리적 관점에서 의견을 제시해 주세요",
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    },
    human_input_mode="NEVER",
)
```

```python
topic = "AI 윤리"

chat_result = topic_agent.initiate_chats(  # 주의 initiate_chat 이 아님
    [
        {
            "recipient": economic_agent,
            "message": f"다음 주제에 대해 경제적 관정에서 의결을 제시해 주세요: {topic}",
            "max_turns": 2,
            "summary_method": "last_msg",
        },
        {
            "recipient": social_agent,
            "message": f"다음 주제에 대해 사회적 관점에서 의결을 제시해 주세요: {topic}",
            "max_turns": 2,
            "summary_method": "last_msg",
        },
        {
            "recipient": environmental_agent,
            "message": f"다음 주제에 대해 환경적 관점에서 의결을 제시해 주세요: {topic}",
            "max_turns": 2,
            "summary_method": "last_msg",
        },
        {
            "recipient": ethical_agent,
            "message": f"다음 주제에 대해 윤리적 관점에서 의결을 제시해 주세요: {topic}",
            "max_turns": 2,
            "summary_method": "last_msg",
        },
    ]
)
```

```python
print("First Chat Summary: ", chat_result[0].summary)
print("Second Chat Summary: ", chat_result[1].summary)
print("Third Chat Summary: ", chat_result[2].summary)
print("Fourth Chat Summary: ", chat_result[3].summary)
```

### 그룹챗 대화방식

```python
from autogen import GroupChat, GroupChatManager

group_chat = GroupChat(
    agents=[topic_agent, economic_agent, social_agent, environmental_agent, ethical_agent],
    messages=[],
    max_round=6,
)

group_chat_manager = GroupChatManager(
    groupchat=group_chat,
    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.getenv("OPENAI_API_KEY"),
            }
        ]
    }
)
```

```python
chat_result = topic_agent.initiate_chat(
    group_chat_manager,
    message="AI 윤리에 대해 대화를 나눠 주세요",
    summary_method="reflection_with_llm",
)
```

```python
print(chat_result.summary)
```
